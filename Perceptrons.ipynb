```python
import numpy as np
import matplotlib.pyplot as plt
import joblib
import pandas as pd
```

```python
class Perceptron:
  def __init__(self,lr,iter):
    self.weights = np.random.randn(3) * 1e-4
    print(f"Weights: {self.weights}")
    self.lr = lr
    self.iter = iter

  def activation_function(self,input,weights):
    z=np.dot(input,weights)
    return np.where(z<0,0,1)

  def fit(self,X,y):
    self.X = X
    self.y = y

    X_with_bias = np.c_[self.X,-np.ones((len(self.X),1))]
    print(f"X with Bias:\n {X_with_bias}")

    for n in range(self.iter):
      y_hat = self.activation_function(X_with_bias,self.weights)
      print(f"Actual Value:\n {self.y}")
      print(f"Predicted Value:\n {y_hat}")

      error = self.y - y_hat
      print(f" Error:\n {error}")

      self.weights = self.weights + self.lr * np.dot(X_with_bias.T,error)
      print(f"Updated weights:\n {self.weights}")
      print("#############\n")

  def predict(self, X):
    X_with_bias = np.c_[X, -np.ones((len(self.X), 1))]
    return self.activationFunction(X_with_bias, self.weights)

```

```python
model = Perceptron(0.5,10)
```
